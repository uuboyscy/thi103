```python
import glob
import pandas as pd
```

```python
"""
data = [
    [
        "推: 24",
        "噓: 2",
        "分數: 22",
        "作者: HTC92 (中国台湾人的驕傲-92共識)",
        "標題: [問卦] PS5已經一年沒開機了  正常嗎？",
        "時間: Sat Jan  6 23:03:29 2024",
    ],
    [
        "推: 24",
        "噓: 2",
        "分數: 22",
        "作者: HTC92 (中国台湾人的驕傲-92共識)",
        "標題: [問卦] PS5已經一年沒開機了  正常嗎？",
        "時間: Sat Jan  6 23:03:29 2024",
    ], ...
]
"""
data = []
for file_path in glob.glob("res_gossiping/*.txt"):
    with open(file_path, "r", encoding="utf-8") as f:
        data.append(
            f.read().split("---split---")[1].split("\n")[1:-1]
        )
data
```

```python
columns = [
    "推",
    "噓",
    "分數",
    "作者",
    "標題",
    "時間",
]

df = pd.DataFrame(data=data, columns=columns)
df
```
